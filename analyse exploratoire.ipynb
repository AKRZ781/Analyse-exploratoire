### Nettoyage des valeurs de la colonne `AGE`
- **Problème identifié** : La colonne `AGE` contient la valeur `"Juvenile"` au lieu d'un âge numérique.
- **Solution** :
  - Remplacement de `"Juvenile"` par `"17"` (valeur par défaut pour les mineurs) avec `data['AGE'].replace()`
  - Vérification des **valeurs uniques** après modification avec `data['AGE'].unique()`
  - Comptage des **valeurs non nulles** avec `data['AGE'].count()`
- **Pourquoi ?** Cela permet d'assurer que toutes les valeurs de `AGE` sont numériques pour faciliter l'analyse.


```python
data['AGE'] = data['AGE'].replace('Juvenile', '17')
data['AGE'].unique()
data['AGE'].count()
```

### Conversion et nettoyage de la colonne `AGE`
- **Objectif** : S'assurer que toutes les valeurs de `AGE` sont numériques pour une analyse correcte.
- **Étapes :**
  1. **Conversion en numérique** :  
     - Utilisation de `pd.to_numeric(data['AGE'], errors='coerce')`  
     - Remplace toutes les **valeurs non numériques** par `NaN`
  2. **Calcul de la moyenne** (`mean_age`) des valeurs valides
  3. **Remplacement des `NaN` par la moyenne** avec `fillna(mean_age)`
  4. **Conversion en entier** (`int`) pour éviter les valeurs décimales
  5. **Affichage des valeurs uniques** après nettoyage


```python
# Convert 'AGE' column to numeric, forcing non-numeric values to NaN
data['AGE'] = pd.to_numeric(data['AGE'], errors='coerce')

# Calculate the mean age
mean_age = data['AGE'].mean()

# Replace NaN values with the mean age and assign it back to 'AGE'
data['AGE'] = data['AGE'].fillna(mean_age)
data['AGE'] =data['AGE'].astype('int')
# Check the unique values in 'AGE' column
print(data['AGE'].unique())
```

###  Évolution du nombre d'arrêts par année
- **Objectif** : Observer la tendance du nombre d'arrêts enregistrés chaque année.
- **Étapes :**
  1. **Extraction de l'année** depuis la colonne `DATETIME` (`.dt.year`)
  2. **Comptage des arrêts par année** avec `.value_counts()`
  3. **Tri chronologique** avec `.sort_index()`
  4. **Affichage sous forme de graphique en ligne** (`kind="line"`)
  5. Ajout de **titres, labels et grille** pour une meilleure lecture
- **Pourquoi ?**  
  - Détecter une **hausse ou une baisse des interventions** au fil des ans.  
  - Comprendre l'impact potentiel de **facteurs externes** (changements de politique, événements, etc.).


```python
# Extraire l'année depuis la colonne DATETIME
data["Year"] = data["DATETIME"].dt.year

# Afficher l'évolution du nombre d'arrêts par année
plt.figure(figsize=(8, 5))
data["Year"].value_counts().sort_index().plot(kind="line", marker='o', title="Nombre d'arrêts par année")
plt.xlabel("Année")
plt.ylabel("Nombre d'arrêts")
plt.grid(True)
plt.show()


```

### Nombre d'arrêts par mois
- **Objectif** : Analyser la distribution des arrêts en fonction du mois de l'année.
- **Étapes :**
  1. Extraction du **mois** à partir de la colonne `DATETIME` avec `.dt.month`
  2. Comptage des arrêts pour chaque mois avec `.groupby("Month")["STOP_TYPE"].count()`
  3. Affichage sous forme de **graphique en ligne** (`kind="line"`)
  4. Ajout de labels et d'une grille pour une meilleure lisibilité
- **Pourquoi ?**  
  - Observer les **pics et creux saisonniers** des interventions.  
  - Identifier d’éventuelles **tendances cycliques** (ex : augmentation des contrôles en été).


```python
data["Month"] = data["DATETIME"].dt.month
plt.figure(figsize=(8, 5))
data.groupby("Month")["STOP_TYPE"].count().plot(kind="line", marker='o', title="Nombre d'arrêts par mois")
plt.xlabel("Mois")
plt.ylabel("Nombre d'arrêts")
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```
